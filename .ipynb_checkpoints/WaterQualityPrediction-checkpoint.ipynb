#### Outlier detection

We're using the Isolation Forest algorithm for outlier detection because it effectively identifies anomalies by isolating observations within random sub-samples of the data. This method is particularly suitable given the complexity and high dimensionality of the dataset, as Isolation Forest is both computationally efficient and robust to irrelevant features.

Due to the complexity of implementing consistent outlier removal within a pipeline—especially with the need to modify both Y_train and Y_test alongside X_train and X_test, we are handling outlier removal outside of the pipeline.

```python
#box plot distribution for features
fig, ax = plt.subplots(ncols=10, nrows=2, figsize=(20,10))
count = 0
ax = ax.flatten()
for i, value in X_train.items():
    sns.boxplot(y=i, data=X_train, ax=ax[count])
    count += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)   
```

```python
# create an instance
iso_model =  IsolationForest(n_estimators=500, max_samples = 0.8, max_features= 0.8, contamination = 0.05, bootstrap=True)

# train isolation forest model
iso_model.fit(X_train)

# predict train outliers
outlier_labels = iso_model.predict(X_train)

# Inlier training dataset
X_train_i = X_train[outlier_labels==1] ## inlier training feature set
Y_train_i = Y_train[outlier_labels==1] ## inlier training target set

# Inlier test dataset
X_test_i =X_test[iso_model.predict(X_test)==1] ## inlier test feature set
Y_test_i =Y_test[iso_model.predict(X_test)==1] ## inlier test target set

# Number of outliers detected and removed in training data
print('Number of outliers removed in training data:',len(X_train[outlier_labels==-1]))
```

#### Building the pipeline

**Scaling**: We are using the StandardScaler to standardize our dataset. Standardization helps ensure that each feature contributes equally to the analysis, improving the performance of subsequent models, especially those sensitive to feature scaling.

**Feature Selection**: We have used ANOVA F-statistic scores and Mutual Information (MI) scores for feature selection to understand the association between features and the target variable. 
A large F-score with a small p-value indicates a strong linear relationship and statistical significance, suggesting that the feature is likely important for prediction. 
A large MI score complements this by capturing the non-linear dependencies. 
This combination helps ensure that selected features are both statistically significant and informative.

**Feature Reduction**: We are using Principal Component Analysis (PCA) to reduce the features in our dataset while retaining 90% of the variance. This method is particularly beneficial because it helps eliminate redundant and collinear features, simplifying the model without losing critical information. By focusing on the most significant components, we can improve the model performance.

```python
# Custom Transformer for Feature Selection
class FeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, f_threshold=0.2, p_threshold=0.001, mi_threshold=0.2):
        self.f_threshold = f_threshold
        self.p_threshold = p_threshold
        self.mi_threshold = mi_threshold

    def fit(self, X, y):
        f, p = f_classif(X, y)
        f /= np.max(f)
        self.mi = mutual_info_classif(X, y)
        self.mi /= np.max(self.mi)

        # visualizing threshold for f score
        plt.figure()
        plt.plot(f, 'o')
        plt.plot(0.2*np.ones(len(f)))
        plt.ylabel('f score')
        plt.xlabel('features')
        
        # visualizing threshold for mi score
        plt.figure()
        plt.plot(self.mi, 'o')
        plt.plot(0.2*np.ones(len(self.mi)))
        plt.ylabel('mi score')
        plt.xlabel('features')

        # create mask
        self.feature_mask = ((f > self.f_threshold) & (p < self.p_threshold)) | (self.mi > self.mi_threshold)
        return self

    def transform(self, X):
        return X[:, self.feature_mask]
```

#### Best Preprocessing Sequence
The preprocessing steps we have chosen are as follows-

1.⁠ ⁠Outlier Detection using Isolation Forest (Implemented outside the pipeline for simplicity) --> 2.⁠ ⁠Scaling using Standard Scaler --> 3. Feature Selection using ANOVA F-statistic scores and Mutual Information (MI) scores --> 4.⁠ ⁠Feature Reduction using PCA

We believe this is the best set of sequence because -

By detecting and removing outliers before scaling, we prevent extreme values from skewing the scaling process, which could disproportionately affect the mean and standard deviation. Scaling is necessary before PCA, as PCA is sensitive to feature scaling. Features with larger magnitudes would dominate the principal components if left unscaled. Implementing feature selection next allows us to retain only the most relevant features based on their relationship with the target variable, which streamlines the dataset for more efficient processing. Feature reduction through PCA is done last, as it helps us reduce dimensionality while retaining the most important variations in the dataset only after the dataset has been cleaned and normalized. These steps should increase the likelihood of enhancing our models' generalization performance.

```python
# Constructing the pipeline
pipeline = Pipeline([
    ('scaling', StandardScaler()),
    ('feature_selection', FeatureSelector(f_threshold=0.2, p_threshold=0.001, mi_threshold=0.2)),
    ('pca', PCA(n_components=0.90))
])
```

```python
# Fitting the pipeline on the training data
X_train_final = pipeline.fit_transform(X_train_i, Y_train_i)

# Applying the pipeline to the test data (without fitting again)
X_test_final = pipeline.transform(X_test_i)
```

```python
# Display shapes of the processed datasets
print(f'Final X_train shape: {X_train_final.shape}')
print(f'Final X_test shape: {X_test_final.shape}')

print(f'\nFinal Y_train shape: {Y_train_i.shape}')
print(f'Final Y_test shape: {Y_test_i.shape}')
```

# 3. Model Training and Testing

## 3.1 Hyperparameter Optimization using Optuna

The 5 models we used are - SVC, Random Forest, Decision tree, Logistic Regression, and KNN

```python
def objective_svc(trial):
    # Define the search space for hyperparameters
    C = trial.suggest_loguniform('C', 0.1, 10.0)
    gamma = trial.suggest_loguniform('gamma', 0.001, 1.0)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf'])
    model = SVC(C=C, gamma=gamma, kernel=kernel, probability=True)
    model.fit(X_train_final, Y_train_i)
    Y_pred = model.predict(X_test_final)
    return accuracy_score(Y_test_i, Y_pred)

def objective_rf(trial):
    # Define the search space for hyperparameters
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_depth = trial.suggest_int('max_depth', 5, 50)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 5)
    
    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf
    )
    model.fit(X_train_final, Y_train_i)
    Y_pred = model.predict(X_test_final)
    return accuracy_score(Y_test_i, Y_pred)

def objective_dt(trial):
    # Define the search space for hyperparameters
    criterion = trial.suggest_categorical('criterion', ['gini', 'entropy'])
    max_depth = trial.suggest_int('max_depth', 1, 20)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 5)
    model = DecisionTreeClassifier(
        criterion=criterion,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf
    )
    model.fit(X_train_final, Y_train_i)
    Y_pred = model.predict(X_test_final)
    return accuracy_score(Y_test_i, Y_pred)

def objective_lr(trial):
    # Define the search space for hyperparameters
    C = trial.suggest_loguniform('C', 0.01, 10.0)
    model = LogisticRegression(C=C)
    model.fit(X_train_final, Y_train_i)
    Y_pred = model.predict(X_test_final)
    return accuracy_score(Y_test_i, Y_pred)

def objective_knn(trial):
    # Define the search space for hyperparameters
    n_neighbors = trial.suggest_int('n_neighbors', 3, 20)
    weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
    model = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
    model.fit(X_train_final, Y_train_i)
    Y_pred = model.predict(X_test_final)
    return accuracy_score(Y_test_i, Y_pred)
```

```python
# Create a function to optimize models and store results
def optimize_and_store_results():
    results = {}

    # SVC Study
    svc_study = optuna.create_study(direction='maximize')
    svc_study.optimize(objective_svc, n_trials=100)
    results['SVC'] = {
        'best_params': svc_study.best_params,
        'best_accuracy': svc_study.best_value
    }

    # Random Forest Study
    rf_study = optuna.create_study(direction='maximize')
    rf_study.optimize(objective_rf, n_trials=100)
    results['Random Forest'] = {
        'best_params': rf_study.best_params,
        'best_accuracy': rf_study.best_value
    }

    # Decision Tree Study
    dt_study = optuna.create_study(direction='maximize')
    dt_study.optimize(objective_dt, n_trials=100)
    results['Decision Tree'] = {
        'best_params': dt_study.best_params,
        'best_accuracy': dt_study.best_value
    }

    # Logistic Regression Study
    lr_study = optuna.create_study(direction='maximize')
    lr_study.optimize(objective_lr, n_trials=100)
    results['Logistic Regression'] = {
        'best_params': lr_study.best_params,
        'best_accuracy': lr_study.best_value
    }

    # KNN Study
    knn_study = optuna.create_study(direction='maximize')
    knn_study.optimize(objective_knn, n_trials=100)
    results['KNN'] = {
        'best_params': knn_study.best_params,
        'best_accuracy': knn_study.best_value
    }

    return results

# Run optimization and get results
results = optimize_and_store_results()
```

```python
# Print results
print("\nOptimized Model Results:\n" + "="*30)
for model_name, result in results.items():
    print(f"{model_name}:")
    print("  Best Parameters:")
    for param, value in result['best_params'].items():
        print(f"    {param}: {value}")
    print("-" * 30)
```

#### Model training and calculating metrics

```python
# Initializing models based on best parameters
models = {
    "SVC": SVC(C=6.4439490614067045, gamma=0.09623789296750944, kernel='rbf', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=111, max_depth=14, min_samples_split=3, min_samples_leaf=1),
    "Decision Tree": DecisionTreeClassifier(criterion='entropy', max_depth=7, min_samples_split=3, min_samples_leaf=2),
    "Logistic Regression": LogisticRegression(C=7.131),
    "KNN": KNeighborsClassifier(n_neighbors=7, weights='distance')
}

# Dictionary to store results
score_results = {}

# Calculate metrics for each model
for model_name, model in models.items():
    model.fit(X_train_final, Y_train_i)
    
    # Memorization (train set)
    Y_train_pred = model.predict(X_train_final)
    train_accuracy = accuracy_score(Y_train_i, Y_train_pred)
    train_f1 = f1_score(Y_train_i, Y_train_pred, average='weighted')
    train_auc = roc_auc_score(Y_train_i, model.predict_proba(X_train_final)[:, 1]) if hasattr(model, "predict_proba") else None
    
    # Generalization (test set)
    Y_test_pred = model.predict(X_test_final)
    test_accuracy = accuracy_score(Y_test_i, Y_test_pred)
    test_f1 = f1_score(Y_test_i, Y_test_pred, average='weighted')
    test_auc = roc_auc_score(Y_test_i, model.predict_proba(X_test_final)[:, 1]) if hasattr(model, "predict_proba") else None
    
    # Store metrics systematically
    results[model_name] = {
        'memorization': {'accuracy': train_accuracy, 'f1_score': train_f1, 'auc': train_auc},
        'generalization': {'accuracy': test_accuracy, 'f1_score': test_f1, 'auc': test_auc}
    }
```

## 3.3 Comparing Memorization & Generalization using F1 score, AUC metrics, and Classification Accuracy

```python
# Function to plot double bar charts with adjusted width and legend position
def plot_metric(metric):
    labels = list(results.keys())
    mem_values = [results[model]['memorization'][metric] for model in labels]
    gen_values = [results[model]['generalization'][metric] for model in labels]
    
    x = np.arange(len(labels))  # Label locations
    width = 0.35  # Bar width
    
    # Increase horizontal length of the plot
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(x - width/2, mem_values, width, label='Memorization', color='skyblue')
    ax.bar(x + width/2, gen_values, width, label='Generalization', color='salmon')
    
    # Labeling
    ax.set_xlabel('Models')
    ax.set_ylabel(metric.capitalize())
    ax.set_title(f'{metric.capitalize()} for Each Model')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    
    # Place legend above the plot to avoid overlap
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2)
    
    plt.show()

# Plot F1 score, accuracy, and AUC
plot_metric('f1_score')
plot_metric('accuracy')
plot_metric('auc')
```

## 4. Impact of Feature Engineering

```python
#Define function to evaluate model performance
def evaluate_model(X_train, X_test, y_train, y_test):
    model = RandomForestClassifier(n_estimators=111, max_depth=14, min_samples_split=3, min_samples_leaf=1)
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    metrics = {
        'test_accuracy': accuracy_score(y_test, y_test_pred),
        'test_f1': f1_score(y_test, y_test_pred),
        'test_auc': roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]),
    }
    return metrics

#  Model performance without polynomial features
metrics = evaluate_model(X_train_final, X_test_final, Y_train_i, Y_test_i)
results = {'Original Features': metrics}
```

```python
# Feature engineering using Polynomial features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train_final)
X_test_poly = poly.transform(X_test_final)
```

Polynomial features refer to the creation of new features by combining existing features through multiplication, resulting in higher-degree polynomial terms. The code uses sklearn's PolynomialFeatures with degree=2, which generates all possible combinations of the original features up to the second degree. For example, if we have features A and B, polynomial features of degree 2 would include A, B, A^2, B^2, and AB. This technique can capture non-linear relationships between features and the target variable, potentially improving the model's ability to fit complex patterns in the data. 
This approach expands the feature space, allowing the model to potentially discover and utilize more complex interactions between variables, which may lead to improved predictive performance.

```python
# Display shapes of the processed datasets
print(f'X_train_poly shape: {X_train_poly.shape}')
print(f'X_test_poly shape: {X_test_poly.shape}')

print(f'\nY_train shape: {Y_train_i.shape}')
print(f'Y_test shape: {Y_test_i.shape}')
```

```python
# Evaluate model performance with polynomial features using the transformed test set
poly_metrics = evaluate_model(X_train_poly, X_test_poly, Y_train_i, Y_test_i)  # Use transformed X_test
results['Engineered Features'] = poly_metrics
```

```python
#Plot results
labels = list(results.keys())
test_accuracies = [result['test_accuracy'] for result in results.values()]
test_f1_scores = [result['test_f1'] for result in results.values()]
test_auc_scores = [result['test_auc'] for result in results.values()]

plt.figure(figsize=(10, 6))
plt.plot(labels, test_accuracies, marker='o', label='Test Accuracy')
plt.plot(labels, test_f1_scores, marker='o', label='Test F1 Score')
plt.plot(labels, test_auc_scores, marker='o', label='Test AUC')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Feature Engineering Steps')
plt.ylabel('Score')
plt.title('Impact of Feature Engineering on Model Generalization Performance')
plt.legend()
plt.tight_layout()
plt.show()
```
